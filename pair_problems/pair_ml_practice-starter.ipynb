```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
import seaborn as sns
```

```python
from sklearn.cross_validation import train_test_split
```

```python
df = pd.read_csv("pair-ml_practice_dataset.csv", index_col=0)
```

```python
df.head(5)
# Note that 'label' is our response variable, i.e., it's our y vector here
```

```python
# Do some EDA
print(df.info())
print(df.describe())
```

```python
## Your Code here
# Some data visualization and plotting here
sns.pairplot(df, x_vars=['feature1','feature2','feature3','feature4'], y_vars='label', size=6, aspect=0.8);
```

```python
sns.pairplot(df, x_vars=['feature1','feature2','feature3','feature4'], y_vars='label', size=6, aspect=0.8, kind='reg');
# In the feature2 plot below we see that there seems to be more of a correlation in the data
```

```python
# Build a Model
# create X and y
feature_cols = ['feature1','feature2','feature3','feature4','feature5','feature6','feature7','feature8','feature9','feature10','feature11','feature12']
X = df[feature_cols]  # We create a matrix X here that holds our features
y = df.label    # We create a vector y that holds our response variables across all the observatons in our dataset.

X,y = df.drop('label', axis=1),df['label']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print(y.head())
print(y_test.head())
print(y_train.head())
print(X.head())
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
print(y.shape)
print(X.shape)
```

```python
# Import Regression Models

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
```

```python
# For each Model Do the following-- we are instatiating the model and fit here
model = LinearRegression()

# fit the model
model.fit(X_train, y_train)

# Predict the model
y_pred=model.predict(X_test)

# Get a Score
print(model.score(X_test,y_test))

# Plot y_test vs y_pred
plt.scatter(y_test, y_pred)
plt.plot([-100,200],[-100,200], 'r--')
print(y_pred)
```

```python
residuals = y_test-y_pred
print(residuals.head(15))  # these are residuals 
plt.scatter(y_pred,residuals)
# You can use this approach on any model
```

```python
# For each Model do the following
# Instantiate the model
model = RandomForestRegressor()

# fit the model
model.fit(X_train, y_train)

# Predict the model
y_pred =  model.predict(X_test)

# Get the score
print(model.score(X_test,y_test))
```

```python
plt.scatter(y_test,y_pred)
plt.plot([-100,200],[-100,200], 'r--')
```

```python
residuals_RF = y_test-y_pred
print(y_test-y_pred)  # these are residuals 
plt.scatter(y_pred,residuals_RF)
# You can use this approach on any model
```

```python
# We can also try to use descition trees using a similar approach as above 
# Import Regression Model
from sklearn.tree import DecisionTreeRegressor

# We are instantiate the model
model = DecisionTreeRegressor(max_depth=2)

# fit the model
model.fit(X_train, y_train)

# Predict the model
y_pred=model.predict(X_test)

# Get a Score
from sklearn.metrics import mean_squared_error, accuracy_score
print(model.score(X_test,y_test))
print("Decision Tree RMSE:",np.sqrt(mean_squared_error(y_test,y_pred)))

# Plot y_test vs y_pred
plt.scatter(y_test, y_pred)
plt.plot([-100,200],[-100,200], 'r--')
```

```python
residuals_RF = y_test-y_pred
print(y_test-y_pred)  # these are residuals 
plt.scatter(y_pred,residuals_RF)
```

```python
from sklearn.grid_search import GridSearchCV, RandomizedSearchCV
# For each Model do the following
# Instantiate the model
model = RandomForestRegressor()
params = {"n_estimators": [10,100,200,300], "min_samples_leaf": [2,5,10,20], "max_features": ["auto","sqrt","log2"] }
grid = RandomizedSearchCV(model, params, n_iter=10, n_jobs=-1)
# fit the model
grid.fit(X_train, y_train)

# Predict the model
y_pred =  grid.predict(X_test)

# Get the score
print(grid.score(X_test,y_test))
# Plot y_test vs y_pred
plt.scatter(y_test, y_pred)
plt.plot([-100,200],[-100,200], 'r--')
```

```python
residuals_RF = y_test-y_pred
print(y_test-y_pred)  # these are residuals 
plt.scatter(y_pred,residuals_RF)
```
